```python
import numpy as np

from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
```

```python
X = data.data
y = data.target

from sklearn.model_selection import ShuffleSplit

ss = ShuffleSplit(n_splits=1, 
                  train_size=0.8, 
                  test_size=0.2, 
                  random_state=0)

train_index, test_index = next(ss.split(X, y))

X_train, X_test = X[train_index], X[test_index]
y_train, y_test = y[train_index], y[test_index]
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
# 平均０，標準偏差１になるようにデータを整形
X_train_scale = scaler.transform(X_train)
```

```python
# 標準化後のデータ
# 平均０，標準偏差１になるようにデータを整形
X_train_scale.mean(axis=0), X_train_scale.std(axis=0)
```

```python
# 標準化前のデータ
X_train.mean(axis=0), X_train.std(axis=0)
```

```python
# 訓練データに対して標準化を行うため、テストデータは厳密に平均０，標準偏差１とはならない
X_test_scale = scaler.transform(X_test)
X_test_scale.mean(axis=0), X_test_scale.std(axis=0) 
```

```python
import matplotlib.pyplot as plt

# スケーリングされているのがよくわかる
plt.scatter(X_train_scale[:, 3], 
            X_train_scale[:, 4], c='blue', 
            label="train")
plt.scatter(X_test_scale[:, 3],  
            X_test_scale[:, 4],  c='red',  
            label="test")
plt.xlabel(data.feature_names[3] + " (standardised)")
plt.ylabel(data.feature_names[4] + " (standardised)")
plt.legend(loc="best");
```

```python
from sklearn import linear_model
clf = linear_model.LogisticRegression()
```

```python
# 標準化したほうが確実に精度があがる
clf.fit(X_train_scale, y_train)
print(clf.score(X_test_scale, y_test))
```

## PCA Whiting
PCA白色化・・・おもに相関関係をなくすために行われる  
具体的にはPCAの処理後に平均０，偏差１とすること

```python
X = data.data[:, [6,7]]
y = data.target
plt.scatter(X[:, 0], X[:, 1])
plt.xlim(0, 0.5)
plt.ylim(0, 0.5)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X)
X_new = pca.transform(X)
plt.scatter(X_new[:, 0], X_new[:, 1])
plt.xlim(-0.1, 0.4)
plt.ylim(-0.25, 0.25)
```

```python
pca = PCA(whiten=True) # PCA白色化を行うOptionを有効化
pca.fit(X)
X_new = pca.transform(X)
plt.scatter(X_new[:,0], X_new[:,1])
plt.xlim(-4, 10)
plt.ylim(-4, 10)
```

## ZCA Whiting

```python
X = np.random.uniform(low=-1, high=1, size=(1000,2)) * (2,1)
y = 2 * X[:,0] + X[:,1]
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlim(-3,3)
plt.ylim(-3,3)
```

```python
angle = np.pi / 4
```

```python
# 回転行列を作成
R = np.array([[np.sin(angle), -np.cos(angle)],
                     [np.cos(angle), np.sin(angle)]])
R
```

```python
# サンプルデータを回転行列に沿って、回転させる
X_rot = X.dot(R)
plt.scatter(X_rot[:, 0], X_rot[:, 1], c=y)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
```

```python
X = X_rot
```

```python
pca = PCA(whiten=False)
pca.fit(X)
X_new = pca.transform(X)
plt.scatter(X_new[:, 0], X_new[:, 1], c=y)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.title("PCA")
```

```python
pca = PCA(whiten=True)
pca.fit(X)
X_new = pca.transform(X)
plt.scatter(X_new[:, 0], X_new[:, 1], c=y)
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.title("PCA Whitening")
```

```python
X_new2 = X_new.dot(pca.components_)
plt.scatter(X_new2[:, 0], X_new2[:, 1], c=y)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.title("ZCA Whiting")
# PCA Whitingを45度回転させたような感じ？
```
